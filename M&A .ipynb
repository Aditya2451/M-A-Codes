```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Vimpel 2011', 'Wind Telecom 2011', 'Vimpel 2011 + Wind Telecom 2011', 'Post Merger Vimpel 2011']
revenue = [20.3, 5.6, 20.3 + 5.6, 21.828]
ebitda = [8.1, 2.1, 8.1 + 2.1, 9.284]

x = np.arange(len(labels))  # label locations
width = 0.35  # bar width

fig, ax = plt.subplots(figsize=(10, 6))

# Plotting
bars1 = ax.bar(x - width/2, revenue, width, label='Revenue (USD Bn)')
bars2 = ax.bar(x + width/2, ebitda, width, label='EBITDA (USD Bn)')

# Labels, title and legend
ax.set_xlabel('Company and Year')
ax.set_ylabel('USD Billion')
ax.set_title('Revenue and EBITDA Comparison')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Display the plot
plt.show()
```
